```python
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
```

```python
words = ['caresses', 'flies', 'dies', 'mules', 'denied',
...            'died', 'agreed', 'owned', 'humbled', 'sized',
...            'meeting', 'stating', 'siezing', 'itemization',
...            'sensational', 'traditional', 'reference', 'colonizer',
...            'plotted']
```

```python
for word  in words:
    print(word+"--->"+stemming.stem(word))
```

```python
for word  in words:
    print(word+"--->"+stemming.stem(word))
```

```python
###RegexpStemmer stemming algo
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer=RegexpStemmer('ing|s$|e$|able$',min=4)
# $ dollar sing dtermines from where the element is to be removed let's say s$ means remove s from the end
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingplaying')
```

```python
###RegexpStemmer Snowball algo
from nltk.stem import SnowballStemmer
```

```python
snowballstemmer=SnowballStemmer('english',ignore_stopwords=False)
```

```python
for word  in words:
    print(word+"--->"+snowballstemmer.stem(word))
```

```python
stemming.stem("fairly"),stemming.stem("sportingly")
```

```python
#better results than poterstemmer in some cases
snowballstemmer.stem("fairly"),snowballstemmer.stem("sportingly")
```

```python
###wordnet lemmatizer
#lematize the words based on Pos -Noun-n Verb-v Adjective a Adverb-r
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
for word  in words:
    print(word+"--->"+lemmatizer.lemmatize(word,pos='v'))
```
