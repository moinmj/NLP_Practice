```python
corpus='Hello welcome to nlp. please read'
```

```python
###paragraph ---> sentence
from nltk.tokenize import sent_tokenize
```

```python
#The output has created 2 differnet sentences from a corpus
sent_tokenize(corpus)
```

```python
#Paragrapgh ---> words
from nltk.tokenize import word_tokenize
```

```python
#the output has enlighted ever single word in the corpus even punctuation
word_tokenize(corpus)
```

```python
#Paragrapgh ---> words
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```
